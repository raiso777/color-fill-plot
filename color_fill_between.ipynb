```python
data = list()
file_name = 'data_T_10'
with open(file_name+'.csv', newline='') as csvfile:
    rows = csv.reader(csvfile)


    for it in rows:
        data.append(it)

upperB = []
meanLine = []
lowerB = []
for i in range(1, len(data)):
    meanLine.append(data[i][1])
    upperB.append(data[i][2])
    lowerB.append(data[i][3])

upperB = np.array(upperB, dtype=float)
meanLine = np.array(meanLine, dtype=float)
lowerB = np.array(lowerB, dtype=float)
x = np.arange(1, len(data))
```

```python
from scipy.interpolate import interp1d
func_upperB = interp1d(x, upperB, kind='cubic')
func_meanLine = interp1d(x, meanLine, kind='cubic')
func_lowerB = interp1d(x, lowerB, kind='cubic')


x = np.linspace(min(x), max(x), 300)
upperB = func_upperB(x)
meanLine = func_meanLine(x)
lowerB = func_lowerB(x)
plt.figure(figsize=(20, 4))
plt.plot(x, upperB, x, lowerB, x,meanLine)
plt.fill_between(x, upperB, lowerB, interpolate=True, color='gray', alpha=0.5)
plt.title('over 90% bin size @maxbin='+file_name.split('_')[2])
plt.ylim([1, 6])
plt.show()
```
